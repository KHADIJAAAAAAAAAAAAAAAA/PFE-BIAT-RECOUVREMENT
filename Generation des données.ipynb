```python
import pandas as pd
import random
from faker import Faker


fake = Faker('fr_FR')


tunisia_first_names = ["Ahmed", "Ali", "Amine", "Anis", "Aymen", "Aziz", "Bilel", "Borhen", "Chaker", "Chamseddine", "Dali", "Dhaker", "Elyes", "Fares", "Fethi", "Foued", "Ghassen", "Hamed", "Hassen", "Haythem", "Iheb", "Ilyes", "Iskander", "Jasser", "Jihen", "Karim", "Khaled", "Lotfi", "Mahdi", "Marwen", "Mehdi", "Mohamed", "Mongi", "Mourad", "Nader", "Nassim", "Nizar", "Omar", "Othman", "Riadh", "Riyad", "Rached", "Rafik", "Sami", "Samiha", "Sarra", "Seif", "Sofiene", "Taoufik", "Tarek", "Walid", "Wissem", "Yacine", "Yassine", "Zied", "Zoubeir", "Abderrahmane", "Abdessalem", "Abdelmajid", "Abdelkader", "Abdelkarim", "Adem", "Aimen", "Akram", "Alaa", "Alaeddine", "Amjed", "Arbi", "Aroua", "Asma", "Atef", "Ayed", "Bahri", "Bassem", "Bechir", "Boujemaa", "Brahim", "Chahinez", "Chems", "Cherif", "Chiraz", "Dalila", "Elyssa", "Emna", "Essia", "Eya", "Fadhel", "Faycel", "Fayza", "Feriel", "Ghofrane", "Ghribi", "Gibril", "Habib", "Hanen", "Hassna", "Hedia", "Helmi", "Hichem", "Houssem", "Ines", "Jamila", "Jemal", "Kaouther", "Kawther", "Lamia", "Lana", "Latifa", "Lilia", "Lina", "Malek", "Mariem", "Mayssa", "Meher", "Mejri", "Melek", "Meriem", "Mina", "Mongi", "Mouadh", "Mourad", "Mouna", "Mounir", "Nabiha", "Nadine", "Najet", "Najla", "Nawfel", "Nedra", "Nejib", "Nermine", "Nesrine", "Nouha", "Olfa", "Ons", "Othmane", "Rabeb", "Rafika", "Rahma", "Rania", "Raouf", "Rebhi", "Reem", "Rim", "Saber", "Sabrine", "Safwen", "Salem", "Samiha", "Sawsen", "Seifeddine", "Selim", "Senda", "Siham", "Sinda", "Sirine", "Sonia", "Soumaya", "Syrine", "Tahar", "Tayeb", "Teycir", "Wael", "Wassila", "Wissem", "Yahya", "Yasmine", "Zahra", "Zeineb", "Zouhir", "Aicha", "Amani", "Amel", "Anwar", "Asfour", "Ateka", "Aya", "Azza", "Badreddine", "Baht", "Baya", "Ben Younes", "Beya", "Boushra", "Chaima", "Chams", "Chekib", "Cyrine", "Dhia", "Dorsaf", "Elhem", "Eman", "Faten", "Fatma", "Faouzi", "Faouzia", "Ferid", "Fouzia", "Ghaya", "Habiba", "Halima", "Henda", "Heykel", "Hosni", "Imen", "Islem", "Jaouhar", "Joud", "Kaies", "Kawther", "Kenza", "Khouloud", "Lamjed", "Lassad", "Lilia", "Lobna", "Mahmoud", "Makram", "Meriem", "Miled", "Moncef", "Mouhib", "Naceur", "Nahla", "Naima", "Nassima", "Nazih", "Nejla", "Nour", "Nouran", "Ouided", "Rahim", "Ramla", "Ranim", "Redha", "Ridha", "Roua", "Rym", "Safa", "Sahbi", "Sakina", "Salha", "Samia", "Sarra", "Selima", "Senda", "Shahd", "Sourour", "Taieb", "Thouraya", "Tijani", "Tir", "Wafa", "Wiem", "Yassmina", "Yosr", "Zayna", "Zina", "Zoubir"]
tunisia_last_names = ["Abassi", "Abdallah", "Abdelkefi", "Abdelmoula", "Abdennadher", "Abidi", "Azaiez", "Azzabi", "Aouinti", "Amri", "Akrout", "Akremi", "Ayari", "Aloui", "Arfaoui", "Arbi", "Arfa", "Ben Ali", "Ben Amor", "Ben Arous", "Ben Attia", "Ben Ayed", "Ben Bouzid", "Ben Cheikh", "Ben Farhat", "Ben Hmida", "Ben Moussa", "Ben Romdhane", "Ben Salah", "Ben Slimane", "Ben Yedder", "Ben Youssef", "Ben Zina", "Bel Hadj", "Bel Haj", "Bel Fkih", "Bel Haj Amor", "Bousnina", "Bouslama", "Bouhlel", "Boughzala", "Bousetta", "Bouguerra", "Boujnah", "Bouali", "Boukthir", "Bougacha", "Boukhari", "Boudabbous", "Boudaya", "Boubaker", "Bouguerra", "Boulabiar", "Bouaziz", "Boujmal", "Boujelbene", "Bouhlel", "Boukhris", "Boumediene", "Bousnina", "Chafai", "Chagra", "Chaouch", "Chatti", "Chafroud", "Chikhaoui", "Cherif", "Chrigui", "Chniti", "Dhouib", "Dhieb", "Dridi", "Dhib", "Dlala", "Doghri", "El Gharbi", "El Hamdi", "El Kamel", "El Mabrouk", "El May", "El Okbi", "Ellouze", "Farhat", "Ferchichi", "Fodha", "Gaaloul", "Gabsi", "Ghariani", "Ghezal", "Ghrab", "Gharsalli", "Gouiaa", "Gouider", "Guesmi", "Guezmir", "Hachicha", "Hadiji", "Hamdi", "Hammami", "Hajji", "Haddad", "Hadj Salem", "Halouani", "Hannachi", "Hantati", "Harzallah", "Hassine", "Hedi", "Hellal", "Hentati", "Houidi", "Jaidi", "Jaziri", "Jelassi", "Jemmali", "Jerbi", "Jeridi", "Jomaa", "Jouini", "Kahlaoui", "Kamoun", "Karray", "Karoui", "Kchok", "Kechrid", "Kechiche", "Kefi", "Khadhar", "Khalfallah", "Khamassi", "Kharraz", "Kharrat", "Khayati", "Khemiri", "Khlif", "Khriji", "Khouini", "Krifa", "Ksibi", "Laarif", "Laaroussi", "Lahmar", "Larbi", "Larnaout", "Lassoued", "Latrach", "Lengliz", "Louati", "Mabrouk", "Machrouh", "Maalej", "Maamouri", "Maaroufi", "Maghrebi", "Mahjoub", "Mahjoubi", "Majdoub", "Makni", "Mallat", "Mansouri", "Marzouk", "Masmoudi", "Mbarek", "Mchirgui", "Meddeb", "Meftah", "Megdiche", "Mellouli", "Melki", "Messaoudi", "Mestiri", "Mezghani", "Mezzi", "Mhamdi", "Mhenni", "Miled", "Mokrani", "Mokhtar", "Mrabet", "Mseddi", "Mzoughi", "Nagati", "Najjar", "Nasri", "Neffati", "Nemri", "Nouri", "Obeid", "Ouali", "Ouertani", "Oueslati", "Ouni", "Rabeh", "Rahali", "Rahem", "Ramzi", "Rassaa", "Rebai", "Redissi", "Rekik", "Riahi", "Romdhani", "Rouissi", "Rzigui", "Saadaoui", "Saafi", "Sabeh", "Sabet", "Saffar", "Sahli", "Sahtout", "Saidi", "Salhi", "Sammoud", "Sassi", "Sayah", "Sekhiri", "Selmi", "Sghaier", "Sghaier", "Sghir", "Sghaier", "Slama", "Slim", "Slimani", "Souli", "Souissi", "Tabib", "Taktak", "Tanfous", "Tarhouni", "Tibaoui", "Tlili", "Touati", "Toumi", "Trabelsi", "Trojet", "Turki", "Yacoub", "Yahiaoui", "Younsi", "Zaag", "Zaafrani", "Zaatour", "Zairi", "Zammouri", "Zarrad", "Zarrouk", "Zayani", "Zebidi", "Zeghidi", "Zehani", "Zitouni", "Zoghlami", "Zribi", "Amdouni", "Ammar", "Aslouj", "Babba", "Bach Tobji", "Bachtaoui", "Bahri", "Bahloul", "Balti", "Barhoumi", "Barkallah", "Barki", "Belajouza", "Belghith", "Bellahmer", "Bellagha", "Belkahla", "Ben Jemaa", "Ben Khelifa", "Ben Nacef", "Ben Nejma", "Ben Souissi", "Ben Zakour", "Berkaoui", "Berrebeh", "Berriri", "Bettaieb", "Boudhina", "Boukhris", "Boulaares", "Boulabiar", "Boumaiza", "Bouzaiene", "Bouzguenda", "Briki", "Bsaies", "Chebbi", "Chouchane", "Chouchen", "Daoulatli", "Debbabi", "Derbel", "Dimassi", "Djebbi", "Djemel", "Driss", "Ezzine", "Fridhi", "Gafsi", "Gharbi", "Ghazi", "Ghedamsi", "Ghezala", "Goucha", "Grira", "Guerfali", "Guesmi", "Hafsi", "Hamrouni", "Hchaichi", "Henchiri", "Jeliti", "Jilani", "Kaabi", "Kacem", "Kadri", "Kallel", "Khaldi", "Khedhiri", "Khiari", "Khmiri", "Labiadh", "Lajmi", "Latrech", "Mabkhout", "Madouri", "Maiza", "Majbouri", "Malouche", "Mannai", "Marai", "Masmoudi", "Mbarek", "Mcharek", "Melliti", "Mkaouar", "Mnasri", "Mounir", "Mzabi", "Nacef", "Naija", "Nemli", "Othmani", "Ouerghi", "Qachoua", "Qardoudi", "Rejeb", "Rekik", "Rezgui", "Rmili", "Romdhane", "Saghrouni", "Sahrane", "Saied", "Safta", "Sakouhi", "Samet", "Selouane", "Sfar", "Sidhom", "Sliti", "Souayah", "Souid", "Taboubi", "Tammar", "Tej", "Tlili", "Trigui", "Younsi", "Zaaraoui", "Zaanouni", "Zaidi", "Zammour", "Zarai", "Zeghlache", "Zitouna", "Zouari", "Zraidi", "Zriba", "Zrig", "Zouaghi"]


tunisia_cities = ["Tunis", "Sfax", "Sousse", "Gabès", "Bizerte", "Ariana", "Nabeul", "Monastir", "Mahdia",  
    "Kairouan", "Gafsa", "Kasserine", "Medenine", "Tataouine", "Tozeur", "Kebili", "Jendouba",  
    "Siliana", "Le Kef", "Sidi Bouzid", "Ben Arous", "Manouba", "Zaghouan", "Beja"]
tunisia_streets = [ 
    "Avenue Habib Bourguiba", "Avenue Mohamed V", "Rue de Marseille", "Avenue de Paris",
    "Boulevard Bab Bhar", "Rue Ibn Khaldoun", "Avenue Farhat Hached", "Rue de la Kasbah",
    "Avenue de Carthage", "Rue du 2 Mars 1934",
    "Avenue Majida Boulila", "Avenue Hédi Chaker", "Rue d’Algérie", "Rue Mongi Slim",
    "Avenue de l’Indépendance", "Boulevard de la République", "Rue du Souk",
    "Boulevard 14 Janvier", "Avenue Yasser Arafat", "Rue des Orangers", "Rue du Ribat",
    "Avenue Abdelhamid Ibn Badis", "Rue d’Italie",
    "Avenue Habib Thameur", "Rue de la Médina", "Rue du Cap Bon", "Avenue Hédi Nouira",
    "Avenue de la Corniche", "Rue de l'Horloge", "Avenue du 7 Novembre", "Rue de la Kasbah",
    "Boulevard de la Mer", "Avenue de la République", "Rue de l’Université", "Avenue Tahar Haddad",
    "Rue des Aghlabides", "Avenue Okba Ibn Nafi", "Rue de la Grande Mosquée", "Avenue Ibn Rochd",
    "Avenue de la Gare", "Rue des Dattes", "Boulevard de l’Environnement",
    "Rue de Houmt Souk", "Avenue Ulysse", "Rue du Port", "Avenue de Midoun"]
tunisian_domains = ["gmail.com", "yahoo.com", "icloud.com"]


def generate_tunisian_name():
    return f"{random.choice(tunisia_first_names)} {random.choice(tunisia_last_names)}"

def generate_tunisian_address():
    return f"{fake.building_number()}, {random.choice(tunisia_streets)} - {random.choice(tunisia_cities)}, Tunisie"

def generate_tunisian_phone():
    return f"+216 {random.randint(20, 99)} {random.randint(100, 999)} {random.randint(100, 999)}"

def generate_tunisian_email(name):
    domain = random.choice(tunisian_domains)
    return f"{name.lower().replace(' ', '.')}_{random.randint(1000, 9999)}@{domain}"

def generate_clients_data(n):
    clients = []
    for _ in range(n):
        name = generate_tunisian_name()
        client = {
            "client_id": fake.unique.random_int(min=1000, max=9999),
            "nom": name,
            "type_client": random.choice(["Particulier", "Entreprise"]),
            "adresse": generate_tunisian_address(),
            "email": generate_tunisian_email(name),
            "telephone": generate_tunisian_phone(),
            "date_creation": fake.date_between(start_date="-5y", end_date="today"),
            "statut_client": random.choice(["Actif", "Inactif"]),
        }
        clients.append(client)
    return pd.DataFrame(clients)

def generate_account_data(clients_df):
    accounts = []
    for _, client in clients_df.iterrows():
        num_accounts = random.randint(1, 3)
        for _ in range(num_accounts):
            account = {
                "account_id": fake.unique.random_int(min=100000, max=999999),
                "client_id": client["client_id"],
                "type_compte": random.choice(["Courant", "Épargne"]),
                "solde": round(random.uniform(0, 50000), 2),
                "date_ouverture": fake.date_between(start_date="-5y", end_date="today"),
                "statut_compte": random.choice(["Actif", "Clôturé"]),
            }
            accounts.append(account)
    return pd.DataFrame(accounts)

def generate_transaction_data(accounts_df):
    transactions = []
    for _, account in accounts_df.iterrows():
        num_transactions = random.randint(5, 20)
        for _ in range(num_transactions):
            transaction = {
                "transaction_id": fake.unique.random_int(min=1000000, max=9999999),
                "account_id": account["account_id"],
                "montant": round(random.uniform(0, 5000), 2),
                "type_transaction": random.choice(["Crédit", "Débit"]),
                "date_transaction": fake.date_between(start_date="-5y", end_date="today"),
                "statut_transaction": random.choice(["Confirmée","En attente","Annulée"]),
            }
            transactions.append(transaction)
    return pd.DataFrame(transactions)

def generate_debts_data(clients_df):
    creances = []
    for _, client in clients_df.iterrows():
        if random.random() < 0.5:  # 50% des clients ont des créances
            num_creances = random.randint(1, 3)
            for _ in range(num_creances):
                creance = {
                    "creance_id": fake.unique.random_int(min=5000, max=99999),
                    "client_id": client["client_id"],
                    "montant_initial": round(random.uniform(500, 20000), 2),
                    "montant_reste_dû": round(random.uniform(500, 20000), 2),
                    "date_echeance": fake.date_between(start_date="today", end_date="+1y"),
                    "statut_creance": random.choice(["En cours","Échue","Récupérée","Contentieuse"]),
                }
                creances.append(creance)
    return pd.DataFrame(creances)

def generate_debt_payments_data(creances_df):
    paiements = []
    for _, creance in creances_df.iterrows():
        if creance["statut"] != "Payée":
            continue
        paiement = {
            "paiement_id": fake.unique.random_int(min=10000, max=99999),
            "creance_id": creance["creance_id"],
            "montant_paye": creance["montant"],
            "date_paiement": fake.date_between(start_date=creance["date_emission"], end_date=creance["date_echeance"]),
            "mode_paiement": random.choice(["Virement","Chèque","Espèces","Prélevement"]),
        }
        paiements.append(paiement)
    return pd.DataFrame(paiements)

N_CLIENTS = 1000  # Nombre de clients

clients_df = generate_clients_data(N_CLIENTS)
accounts_df = generate_account_data(clients_df)
transactions_df = generate_transaction_data(accounts_df)
debts_df = generate_debts_data(clients_df)
payments_df = generate_debt_payments_data(debts_df)

with pd.ExcelWriter('recouvrement_data.xlsx', engine='openpyxl') as writer:
    clients_df.to_excel(writer, sheet_name='CLIENT_OP', index=False)
    accounts_df.to_excel(writer, sheet_name='COMPTE_OP', index=False)
    transactions_df.to_excel(writer, sheet_name='TRANSACTION_OP', index=False)
    debts_df.to_excel(writer, sheet_name='CREANCE_OP', index=False)
    payments_df.to_excel(writer, sheet_name='PAIEMENT_OP', index=False)

print("Données enregistrées dans recouvrement_data.xlsx avec plusieurs feuilles.")
```
